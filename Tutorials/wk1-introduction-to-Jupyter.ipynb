### The Notebook has Command and Edit modes.

These are the command and edit modes of your notebook.

In command mode, pressing the “h” key will provide a list of all the shortcut keys.

Command mode alows you to edit notebook-level features, and edit mode changes the content of cells.

    The “b” key will make a new cell below the currently selected cell.
    The “a” key will make one above.
    The “x” key will delete the current cell.
    
Use __shift+enter__ in command or edit mode to run cells.

### Slides

You can turn your Jupyter notebook into a nice slideshow which can be very useful for presentations! You can even skip cells or have optional cells to show! To turn your notebook into a slideshow, follow the next instructions. Look into the upper menu:

    View --> Cell Toolbar --> Slideshow

Now every cell with have different options. You can configurate them however you what to create your slideshow.

Run this command in you terminal to turn your notebook into a slideshow afterwards:
    
    jupyter nbconvert Slides.ipynb --to slides --post serve
